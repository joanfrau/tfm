# Pitch Plots

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.patches as mpatches
import matplotlib.pyplot as pl

from mplsoccer import Pitch, VerticalPitch
```

```python
pitch1 = VerticalPitch(pitch_color='grass', line_color='white',
              stripe=False,goal_type='box')
fig1, ax1 = pitch1.draw(figsize=(25, 14),nrows=1, ncols=5)

#campo 1
a = ax1[0].plot(40,110,'ro', markersize=5)
a = ax1[0].plot(40,118,'bo', markersize=5)

#campo 2
a = ax1[1].plot(40,110,'ro', markersize=5)
a = ax1[1].plot(40,118,'bo', markersize=5)

#campo 3
a = ax1[2].plot(35,85,'ro', markersize=5)
a = ax1[2].plot(40,116,'bo', markersize=5)
a = ax1[2].plot(41,105,'bo', markersize=5)
a = ax1[2].plot(36,100,'bo', markersize=5)

#campo 4
a = ax1[3].plot(30,100,'ro', markersize=5)
a = ax1[3].plot(36,112,'bo', markersize=5)

#campo 5
a = ax1[4].plot(40,102,'ro', markersize=5)
a = ax1[4].plot(40,117,'bo', markersize=5)
a = ax1[4].plot(41,104,'bo', markersize=5)
```

```python
dataset = pd.read_csv("data/all_shots.csv")
dataset["goal"] = dataset.apply(lambda row: 1 if row.outcome=="Goal" else 0,axis=1)

pitch4 = VerticalPitch(pitch_color='grass',line_color='white',goal_type='box')
fig4, ax4 = pitch4.draw(figsize=(8, 16))
for i in dataset.iterrows():
    i=i[1]
    if i["goal"] == 1:
        a = ax4.plot(i["start_location_y"],i["start_location_x"],'bo', markersize=0.5)
    else:
        a = ax4.plot(i["start_location_y"],i["start_location_x"],'ro', markersize=0.5)

red_patch = mpatches.Patch(color='red', label='No goal')
blue_patch = mpatches.Patch(color='blue', label='Goal')
plt.legend(handles=[red_patch, blue_patch],loc="lower right")
```

```python
dataset = dataset[dataset["player"]=="Lionel Andrs Messi Cuccittini"]
pitch5 = VerticalPitch(pitch_color='grass',line_color='white',goal_type='box')
fig5, ax5 = pitch5.draw(figsize=(8, 16))
for i in dataset.iterrows():
    i=i[1]
    if i["goal"] == 1:
        a = ax5.plot(i["start_location_y"],i["start_location_x"],'bo', markersize=1)
    else:
        a = ax5.plot(i["start_location_y"],i["start_location_x"],'ro', markersize=1)

red_patch = mpatches.Patch(color='red', label='No goal')
blue_patch = mpatches.Patch(color='blue', label='Goal')
plt.legend(handles=[red_patch, blue_patch],loc="lower right")
```
